```python
def path_finder(list_path, string):
    list_path_new = []
    for path in list_path:
        if os.path.isdir(path):
            with os.scandir(path) as entries:
                for item in entries:
                    if string =="":
                        list_path_new.append(item.path)
                    elif item.is_dir() and re.search(string, item.name, re.I):
#                         print(item.path)
                        list_path_new.append(item.path)
    return list_path_new
```

```python
def hitory_finder(list_path):
    list_path_history = []
    for path in list_path:
        if os.path.isdir(path):
            list_file = os.listdir(path)
            for file in list_file:
                if file == "History.txt" :
                    list_path_history.append(os.path.join(path, file))
                    
    return list_path_history
```

```python
def find_dir(path):
    # 函數功能: 遞迴顯示指定路徑下的所有檔案及資料夾名稱
    for fd in os.listdir(path):
        full_path=os.path.join(path,fd)
        if os.path.isdir(full_path):
            print('資料夾:',full_path)
            find_dir(full_path)
        else:
            print('檔案:',full_path)
find_dir("./")
```
