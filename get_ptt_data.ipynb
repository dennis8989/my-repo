```python
import requests as rs

def geturl(URL):
    res = rs.get(URL,cookies={'over18': '1'})
    soup = BeautifulSoup(res.text,'html.parser')
    btn = soup.find_all(class_="btn wide")
    page = int(re.search('\d{5}',btn[1]['href']).group())+1


    for i in range(2):
        urls_new = "https://www.ptt.cc/bbs/Gossiping/index" + str(page - i) + ".html"
        yield urls_new 
```

```python
def filter_data(urls_new):
    for url_new in urls_new:
        print(url_new)
        res = rs.get(url_new,cookies={'over18': '1'})
        soup = BeautifulSoup(res.text,'html.parser')
        soup_div = soup.find_all('div',class_="r-ent")
        length = len(soup_div)
        # soup_div[0]
        for i in range(length):
            soup_div[i].find(class_="date").text.strip()
            soup_div[i].find('a').text
            soup_div[i].find(class_='author').text
            soup_div[i].find('a')['href']
            try:
                soup_div[i].find('span').text
            except :
                print(0)
```

```python
URL = 'https://www.ptt.cc/bbs/Gossiping/index.html'
urls_new = geturl(URL)
filter_data(urls_new)
# for url_new in urls_new:
#     print(url_new)
```
